```python
d3 = pandas.read_csv('./Standard/' + STATE_CODE + '_LOCALITY_psv.psv', delimiter='|')
```

## Join

```python
join1 = d2.merge(right=d1, how='left', on='ADDRESS_DETAIL_PID')
```

```python
join2 = join1.merge(right=d3, how='left', on='LOCALITY_PID')
```

```python
keep =  [
    'POSTCODE',
    'LOCALITY_PID',
    'LOCALITY_NAME',
    'LONGITUDE',
    'LATITUDE'
]
```

```python
subset = join2[keep]
```

```python
subset.POSTCODE.nunique()
```

```python
subset.LOCALITY_NAME.nunique()
```

```python
localities = subset.LOCALITY_NAME.unique()
```

```python
data = [{'LOCALITY_NAME': l} for l in localities]
```

```python
for (index, item) in enumerate(data):
    if (index % 10):
        print(index, len(data))
    sub = subset[subset.LOCALITY_NAME == item['LOCALITY_NAME']]
    POSTCODE = sub.POSTCODE.min()
    LONGITUDE = sub.LONGITUDE.mean()
    LATITUDE = sub.LATITUDE.mean()
    data[index]['POSTCODE'] = POSTCODE
    data[index]['LONGITUDE'] = LONGITUDE
    data[index]['LATITUDE'] = LATITUDE
    data[index]['STATE_CODE'] = STATE_CODE
```

## Export results

```python
filename = STATE_CODE + '.json'
```

```python
filename
```

```python
dataframe = pandas.DataFrame(data)
```
